```python
import requests
from bs4 import BeautifulSoup
```

```python
BASE_URL = "http://search.naver.com/search.naver?where=post&sm=tab_jum&ie=utf8&query=%EC%95%84%EC%9D%B4%EC%98%A4%EC%95%84%EC%9D%B4"
```

```python
response = requests.get(BASE_URL)
```

```python
response.headers
```

```python
response.status_code # 2xx => successful
                     # 3xx => redirection
                     # 4xx => failed
        
        
```

```python
response.content  # text는 str 타입으로 나옴
```

```python
dom = BeautifulSoup(response.content, "html.parser") # 클래스의 객체 생성
```

```python
# dom => Document Object Model : 문서 객체
# html > div > h1 > p > ul > li > .... 이런 하나를 객체로 보는 것
```

```python
# dom
```

```python
# parsing

# tag => 태그 이름( "div", "a")
# id => 문서 전체에서 유일한 unique 값  => "#"
# class => 문서 전체에서 1-n개. => 공통적인 특징을 가진 tag들 => "."
```

```python
post_elements = dom.select("li.sh_blog_top")
```

```python
post_element = post_elements[0]

title_element = post_element.select_one("a.sh_blog_title")
```

```python
title_element
```

```python
title_element.text
```

```python
title_element.get("href")
```

```python
for post_element in post_elements:
    title_element = post_element.select_one("a.sh_blog_title")
    title = title_element.text
    url = title_element.get("href")
    print((title, url))
```

```python
for room_id in range(5667010,5667010+10):
    BASE_URL = "https://api.zigbang.com/v1/items?detail=true&item_ids=" + str(room_id)
    response = requests.get(BASE_URL)
    room_information = json.loads(response.text)
    
    try:
        address = room_information["items"][0]["item"]["agent_address1"]
        deposit = room_information["items"][0]["item"]["deposit"]
        rent = room_information["items"][0]["item"]["rent"]
        
        df.loc[len(df)] = [address, int(deposit), int(rent)]
        
        print((address, deposit, rent))
    except:
        pass
    
```
